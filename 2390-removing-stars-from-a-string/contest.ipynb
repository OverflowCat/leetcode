```python
def removeStars(s: str) -> str:
    st = ord('*')
    s = bytearray(s, 'ascii')
    l = len(s)
    maxnonsp = -1
    for i in range(l):
        if s[i] == st:
            if maxnonsp != -1 and s[maxnonsp] != st:
                s[maxnonsp] = st
                maxnonsp -= 1
            else:
                j = i - 1
                if maxnonsp == -1:
                    j = maxnonsp
                while s[j] == st:
                    j -= 1
                s[j] = st
                maxnonsp = j - 1
        else:
            maxnonsp = i
    return s.decode('ascii').replace('*', '')
```
